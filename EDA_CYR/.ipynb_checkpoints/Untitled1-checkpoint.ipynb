```python
import os
path = os.path.dirname(os.getcwd())
path = os.path.join(path, "ProcessedData")
train = pd.read_csv(os.path.join(path,"merged_train.csv"))
test = pd.read_csv(os.path.join(path,"merged_test.csv"))
```

```python
train_one = pd.read_csv("data/train2.csv")
test_one = pd.read_csv("data/test2.csv")
result = pd.read_csv("data/result.csv")
```

```python
for i in ['min','25%','50%','75%','max']:
    for j in ['임대료','임대보증금']:
        train_one = train_one.drop('store'+j+i,axis=1)
        
train_one = train_one.drop(['sub_0','sub_1'],axis=1)

for i in [0,1,2,3,8,9,10,11,12,13]:
    train_one = train_one.drop('type_'+str(i),axis=1)
    
for i in [0,1,2]:
    train_one = train_one.drop('sub_dist_'+str(i),axis=1)
    
train_one = train_one.drop(['지하철역','버스정류장'],axis=1)
train_one = train_one.drop(['store전용면적mean','store전용면적std'],axis=1)

train_one = train_one.drop(['연면적'],axis=1)

train['남자'] = train.apply(lambda x : x['총세대수']/(x['남/여비율']+1),axis=1)
tmp = train[['단지코드','남자']].drop_duplicates().reset_index(drop=True)
train_one['남자'] = tmp['남자']

train['가구수'] = train.apply(lambda x : x['총세대수']/(x['세대당_인구']+1),axis=1)
tmp = train[['단지코드','가구수']].drop_duplicates().reset_index(drop=True)
train_one['가구수'] = tmp['가구수']
```

```python
def func3(x,dic):
    for i in dic.keys():
        if len(dic[i]) == len(x):
            if all(dic[i] == x):
                return i


train.loc[train['공급유형'].isin(['영구임대', '행복주택']),'공급유형'] = '영구/행복'
test.loc[test['공급유형'].isin(['영구임대', '행복주택']),'공급유형'] = '영구/행복'

train.loc[train['공급유형'].isin(['공공임대(10년)', '공공임대(5년)']),'공급유형'] = '공공임대(단기)'
test.loc[test['공급유형'].isin(['공공임대(10년)', '공공임대(5년)']),'공급유형'] = '공공임대(단기)'

no_shop = train[train['공급유형']!='임대상가']
uniq = np.reshape(no_shop['공급유형'].unique(),(7,1))

dic = dict(enumerate(uniq))
dic[8] = ['공공임대(단기)', '국민임대']
dic[9] = ['장기전세', '국민임대', '영구/행복']
dic[10] = ['국민임대', '영구/행복']
dic[11] = ['공공임대(단기)', '공공임대(분납)']
dic[12] = ['공공분양', '공공임대(단기)', '공공임대(분납)']
dic[13] = ['공공임대(50년)', '영구/행복']
dic[14] = ['장기전세', '공공임대(단기)', '공공임대(분납)']

dic1 = dict()
for i in dic.keys():
    val = train.loc[train['공급유형'].isin(dic[i]),'임대료'].mean()
    dic1[i] = val


no_shop = train[train['공급유형']!='임대상가']
sample = pd.DataFrame(no_shop.groupby('단지코드')['공급유형'].unique()).reset_index()
sample['cluster'] = sample['공급유형'].apply(lambda x : func3(x,dic))
sample['cluster'] = sample['cluster'].apply(lambda x: int(x))
train = pd.merge(train,sample[['단지코드','cluster']],on='단지코드',how='left')
train['cluster_임대료'] = train['cluster'].apply(lambda x: dic1[x])
train = train.drop(['공급유형'],axis=1)


no_shop = test[test['공급유형']!='임대상가']
sample = pd.DataFrame(no_shop.groupby('단지코드')['공급유형'].unique()).reset_index()
sample['cluster'] = sample['공급유형'].apply(lambda x : func3(x,dic))
test = pd.merge(test,sample[['단지코드','cluster']],on='단지코드',how='left')
test['cluster_임대료'] = test['cluster'].apply(lambda x: dic1[x])
test = test.drop(['공급유형'],axis=1)
```

```python
X = train_one.drop(['등록차량수','단지코드','cluster_임대보증금'],axis=1)
y = train_one['등록차량수']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=2021,shuffle=True)

cb = CatBoostRegressor(random_state=2021,verbose=False,loss_function='MAE').fit(X_train, y_train)
pred = cb.predict(X_test)

mae(pred,y_test) 
```
